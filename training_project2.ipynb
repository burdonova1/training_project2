```python
from selenium.webdriver import Chrome
import chromedriver_binary
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import time
import csv
import pandas as pd
```

```python
cvsFile=open("amazonProducts.csv", "wt")
writer=csv.writer(cvsFile)

driver=Chrome()
driver.get("https://www.amazon.com/")
box=driver.find_element(By.ID, "twotabsearchtextbox")
box.send_keys("air Fryer")
box.send_keys(Keys.ENTER)
time.sleep(3)
cards=driver.find_elements(By.CLASS_NAME, "s-card-container")
titles=[]
prices=[]

for card in cards:
    try:
        title = card.find_element(By.CLASS_NAME, "a-size-medium").text
        price=card.find_element(By.CLASS_NAME, "a-price").text
        clean_price=price.replace("\n",".").strip("$")
 #       print(price,title)
        alist=[title, clean_price]
        titles.append(title)
        prices.append(clean_price)
 #       writer.writerow(alist)
    except: 
        pass
#cvsFile.close()
#driver.close()
df=pd.DataFrame()
df["title"]=titles
df["price"]=prices
df["price"]=pd.to_numeric(df["price"])
min_price=df["price"].min()
condition=df["price"]==min_price
df[condition]

    
```
